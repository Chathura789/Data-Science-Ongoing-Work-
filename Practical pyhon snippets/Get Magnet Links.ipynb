```python
def get_magnet_links(content):
    pq = pyquery.PyQuery(content)    
    seen = set()
    for a in pq.find('a'):
        href = a.get('href')
        if href and (href.startswith("magnet")):
                seen.add(href)
    return seen
```

```python
link ='https://rarbgunblock.com/torrent/y24nex9'
```

```python
get_magnet_links(link)
```
