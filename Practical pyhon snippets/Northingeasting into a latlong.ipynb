```python
escapes = pd.read_csv('https://www.dropbox.com/s/skmtmf2k13lorfz/escapes.csv?dl=1',encoding='cp1252')
escapes.head()
```

```python

def proj_transform(df):
    bng = pyproj.Proj("+init=EPSG:27700")
    wgs84 = pyproj.Proj("+init=EPSG:4326")
    lats = []
    lons = []
    for k, val in enumerate(df['Easting']):
        lon, lat = pyproj.transform(bng,wgs84,df['Easting'][k], df['Northing'][k])
        lats.append(lat)
        lons.append(lon)
    df['lat'] = lats
    df['lon'] = lons
    return df


```

```python
df_transformed = proj_transform(escapes)

df_transformed.head()
```
