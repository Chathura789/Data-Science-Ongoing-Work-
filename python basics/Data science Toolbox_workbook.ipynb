```python
# now write a function called reaise both 
# eg : x raised to y and y raised to x 

def raiseboth(x,y):
  value_01 = x**y 
  value_02 = y**x
  new_tuple = (value_01,value_02) 
  return new_tuple 

```

```python
import pandas as pd 
df = pd.read_csv('Superstore_Orders.csv')
```

```python
# Initialize an empty dictionary: langs_count
item_count = {}

# Extract column from DataFrame: col
col = df['Manufacturer']

# Iterate over Postal Code column in dataframe
for entry in col:

    # If the language is in langs_count, add 1
    if entry in item_count.keys():
        item_count[entry] += 1
    # Else add the language to langs_count, set the value to 1
    else:
        item_count[entry] = 1

# Print the populated dictionary
print(item_count)
```

```python
# check 
# Define count_entries()
def count_entries(df, col_name):
    """Return a dictionary with counts of 
    occurrences as value for each key."""

    # Initialize an empty dictionary: langs_count
    langs_count = {}
    
    # Extract column from DataFrame: col
    col = df[col_name]

    # Iterate over lang column in dataframe
    for entry in col:

        # If the language is in langs_count, add 1
        if entry in langs_count.keys():
            langs_count[entry] += 1
        # Else add the language to langs_count, set the value to 1
        else:
            langs_count[entry] = 1

    # Return the langs_count dictionary
    return langs_count

# Call count_entries(): result
result = count_entries(df, 'lang')

# Print the result
print(result)
```
