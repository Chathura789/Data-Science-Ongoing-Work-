```python
print("=========================================================")
print("====-     Bank Information Management System     -====")
print("=========================================================\n")

file = open("list.txt","r")

content = file.readlines()

student_list=[]

for student_id in content:
    student_list.append(student_id)

file.close()


while True:
    print("-------------------------------------")
    print("Things you can do : ")
    print("    1 . View current accounts")
    print("    2 . Add a new customer account")
    print("    3 . Remove a customer account")
    print("    4 . Deposit cash")
    print("    5 . Withdraw cash")
    print("    6 . Exit")
    print("-------------------------------------")
    
    activity = input("\nPlease enter the activity number : ")

    if activity=="6":
        break

    if activity=="1":
        print("\n Current Customer List")
        for std_id in student_list:
            print(" "+std_id.strip())
        print()


    if activity=="2":
        std_id=input("\n  Please enter Customer  : ")
        initdep=input("Please enter initial amount to deposit : ")
        student_list.append(std_id +","+ initdep)
        

    if activity=="3":
        std_id=input("\n  Please enter Customer ID and total amount remaining to delete\n" "- example(name,money) : ")
        student_list.remove(std_id)

    if activity == "4":
        x=input("\n please choose an account : ")
        deposit=int(input("Please Enter the amount to deposit : "))
        for std_id in content:
                chunks=std_id.split(",")
                print(chunks[0])
                if x == chunks[0]:
                    
                    if deposit == chunks[0]:
                        student_list.append(deposit+","+int(chunks[1]))
                        
    if activity == "5":
        x=input("\n Please enter the account and current balance to withdraw : ")
        withdraw=int(input("Please enter the amount to withdraw : "))
        for std_id in content:
            chunks=std_id.split(",")
            #print(chunks[0])
            if x == chunks[0]:#Identifying the correct user
                current_amount=chunks[1]
                student_list.remove(x)
                new_amount = current_amount-withdraw
                student_list.append(x+ ","+new_amount)
                #
                #if withdraw == chunks[1]:
                    #student_list.remove(withdraw-","-int(chunks[1]))
        

    input("Press any key to continue...")

##Writing changes to a file
print("---   Saving changes to the file   ---")
file = open("list.txt","w+")

for std_id in student_list:
    file.write(std_id)

file.close()
print("---   All changes saved sucessfully   ---")


print("\n    Have a nice day .. !\n")
```
